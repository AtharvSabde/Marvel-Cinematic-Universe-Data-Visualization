## Movies Per Year

```python
film_year = df.Year_Release.value_counts().rename_axis('Year_Release').reset_index(name='Film') 
plt.figure(figsize=(14,8))
plt.title("No. of Films released each year")
c1 = sns.barplot(x="Year_Release",y="Film",data=film_year,palette="flare")
c1.bar_label(c1.containers[0])
plt.show()
```

## Box office gross Worldwide

```python
box_ofc_p1 = df.query("Phase=='Phase One'")
box_ofc_p2 = df.query("Phase=='Phase Two'")
box_ofc_p3 = df.query("Phase=='Phase Three'")
box_ofc_p4 = df.query("Phase=='Phase Four'")
box_ofc_p5 = df.query("Phase=='Phase Five'")

trace1 = go.Bar(
    x=box_ofc_p1['Film'],      
    y=box_ofc_p1['Box office gross Worldwide'],
    name = 'Phase 1',
    marker=dict(color='#FFD700')
)
trace2 = go.Bar(
    x=box_ofc_p2['Film'],      
    y=box_ofc_p2['Box office gross Worldwide'],
    name = 'Phase 2',
    marker=dict(color='#9EA0A1')
)
trace3 = go.Bar(
    x=box_ofc_p3['Film'],      
    y=box_ofc_p3['Box office gross Worldwide'],
    name = 'Phase 3',
    marker=dict(color='#CD7F32')
)
trace4 = go.Bar(
    x=box_ofc_p4['Film'],      
    y=box_ofc_p4['Box office gross Worldwide'],
    name = 'Phase 4',
    marker=dict(color='#0b7314')
)
trace5 = go.Bar(
     x=box_ofc_p5['Film'],
     y=box_ofc_p5['Box office gross Worldwide'],
     name = 'Phase 5',
     marker = dict(color = '#f01707')
)
data = [trace1, trace2, trace3, trace4, trace5]
layout = go.Layout(
    title='Worldwide Box Office by Phases'
)
fig = go.Figure(data=data, layout=layout)
fig.show()
```

## Films And Their Budget

```python
budget_rate = df.groupby('Film').mean().sort_values('Budget', ascending = False).index

plt.figure(figsize=(20,12))
ax = sns.barplot(y = "Film", x = "Budget" , data = df, order = budget_rate)
ax.bar_label(ax.containers[0])
plt.title(" Film with Highest Budget" , size = 20)
plt.show()
plt.clf()
```

## All-time ranking Worldwide

```python
rating = df.groupby('Film').mean().sort_values('All-time ranking Worldwide', ascending = True).index

plt.figure(figsize=(20,12))
ax = sns.barplot(y = "Film", x = "All-time ranking Worldwide" , data = df, order = rating)
ax.bar_label(ax.containers[0])
plt.title("All-time ranking Worldwide" , size = 20)
plt.show()
plt.clf()
```

## Rotten Tomatoes Rating

```python
order_rate = df.groupby('Film').mean().sort_values('Rotten Tomatoes Rating', ascending = False).index

plt.figure(figsize=(20,12))
ax = sns.barplot(y = "Film", x = "Rotten Tomatoes Rating" , data = df, order = order_rate)
ax.bar_label(ax.containers[0])
plt.title(" Film with Highest Rotten Tomatoes Rating" , size = 20)
plt.show()
plt.clf()
```
